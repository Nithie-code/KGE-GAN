```python
import numpy as np

# Number of simulations
n = 1000000

# Step 1: Break the stick at a random point between 0 and 1
break1 = np.random.rand(n)

# Step 2: Break the longer piece again
longer_piece = np.maximum(break1, 1 - break1)
break2 = np.random.rand(n) * longer_piece

# Get the lengths of the three pieces
piece1 = np.minimum(break1, 1 - break1)
piece2 = break2
piece3 = longer_piece - break2

# Check the triangle inequality conditions
valid_triangles = (piece1 + piece2 > piece3) & (piece1 + piece3 > piece2) & (piece2 + piece3 > piece1)

# Calculate the probability
probability = np.mean(valid_triangles)

print(f"Probability that the pieces form a triangle: {probability:.5f}")
```
