```python
import matplotlib.pyplot as plt
import numpy as np

# X values from 5 to 25 with increments of 5
x = np.arange(5, 20, 5)  # 5, 10, 15, 20, 25

# Y values from 0.40 to 0.80 with increments of 0.05
y = np.arange(0.40, 0.95, 0.05)
x = np.arange(5, 20, 5)
# Sample data for the line chart
GRSLA=[]
CRHITS=[]
BNR=[]
RARSB=[]
GCRGAN=[]
SCRNTR=[]
KGETransD=[]
KGEGAN=[]

X= [5,10,15,20]  # Y-axis values

# Create a line chart with multiple x-variables
plt.plot(X,KGEGAN, marker='s', linestyle='-', color='m', label='KGE-GAN')
plt.plot(X,GCRGAN, marker='s', linestyle='-', color='y', label='GCR-GAN')
plt.plot(X,KGETransD, marker='s', linestyle='-', color='g', label='KGE-TransD')
plt.plot(X,CRHITS,marker='s', linestyle='-', color='b', label='CR-HITS')
plt.plot(X,RARSB,marker='s', linestyle='-', color='black', label='RAR-SB')
plt.plot(X,SCRNTR,marker='s', linestyle='-', color='red', label='SCR-NTR')
plt.plot(X,BNR, marker='s', linestyle='-', color='brown', label='BNR')
plt.plot(X,GRSLA, marker='s', linestyle='-', color='c', label='GRSLA')

# Add titles and labels
#plt.title('Line Chart with Multiple X Variables')
plt.xlabel('Number of K',size=16)
plt.ylabel('Recall',size=16)
plt.xlim(2.5, 21)
plt.ylim(0.40, 0.85)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xticks([5, 10, 15, 20], fontsize=14)  # Set specific tick labels
# Display the legend
plt.legend()
plt.savefig("Recall-DBLP-V12.pdf", format='pdf')
# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt

# Sample data for the line chart
x1 = [1, 2, 3, 4, 5]  # X1 values
x2 = [2, 3, 4, 5, 6]  # X2 values
x3 = [3, 4, 5, 6, 7]  # X3 values
x4 = [4, 5, 6, 7, 8]  # X4 values
x5 = [5, 6, 7, 8, 9]  # X5 values
x6 = [6, 7, 8, 9, 10] # X6 values
y = [10, 20, 15, 25, 30]  # Y-axis values

# Create a line chart with research paper style colors and markers
plt.plot(x1, y, marker='o', linestyle='-', color='k', label='X1 Data')  # Black circle
plt.plot(x2, y, marker='s', linestyle='-', color='k', label='X2 Data')  # Dark gray square
plt.plot(x3, y, marker='^', linestyle='-', color='b', label='X3 Data')  # Blue triangle up
plt.plot(x4, y, marker='d', linestyle='-', color='r', label='X4 Data')  # Red thin diamond
plt.plot(x5, y, marker='*', linestyle='-', color='g', label='X5 Data')  # Green star
plt.plot(x6, y, marker='x', linestyle='-', color='k', label='X6 Data')  # Dark gray cross

# Add titles and labels
plt.title('Line Chart with Research Paper Colors and Markers')
plt.xlabel('X-axis Values')
plt.ylabel('Y-axis Values')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```
